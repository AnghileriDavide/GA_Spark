```python
class Mutation:
    
    _strategy = {
        'rand1' : '_rand_1', 
        'rand2' : '_rand_2',
        'best1' : '_best_1', 
        'best2' : '_best_2',
        'target_best': '_target_best'
    }
    
    def __init__(self, f, MuSt = "rand1"):
        self.f = f
        self.mutate = getattr(self, self._strategy[MuSt])
        
        
    def _sorted(self, population):
        return sorted(population, key = lambda x: x[1][1])
    
    
    def _random_indv(self, pop_len, size, best = False):
        return random.sample(range(int(best), pop_len), size)
    
    
    def _rand_1(self, population, target = None):
        indv = self._random_indv(len(population), 3)
        
        a = population[indv[0]][1][0]
        b = population[indv[1]][1][0]
        c = population[indv[2]][1][0]
        
        return a + self.f * (b - c)
    
    
    def _rand_2(self, population, target = None):
        indv = self._random_indv(len(population), 5)
        
        a = population[indv[0]][1][0]
        b = population[indv[1]][1][0]
        c = population[indv[2]][1][0]
        d = population[indv[3]][1][0]
        e = population[indv[4]][1][0]
        
        return a + self.f * (b - c) + self.f * (d - e)
    
    
    def _best_1(self, population, target = None):
        sorted_pop = self._sorted(population)
        indv = self._random_indv(len(sorted_pop), 2, True)
        
        best = sorted_pop[0][1][0]
        a = sorted_pop[indv[0]][1][0]
        b = sorted_pop[indv[1]][1][0]
        
        return best + self.f * (a - b)
    
    
    def _best_2(self, population, target = None):
        sorted_pop = self._sorted(population)
        indv = self._random_indv(len(sorted_pop), 4, True)
        
        best = sorted_pop[0][1][0]
        a = sorted_pop[indv[0]][1][0]
        b = sorted_pop[indv[1]][1][0]
        c = sorted_pop[indv[2]][1][0]
        d = sorted_pop[indv[3]][1][0]
        
        return best + self.f * (a - b) + self.f * (c - d)
    
    
    def _target_best(self, population, target):
        sorted_pop = self._sorted(population)
        best = sorted_pop[0][1][0]
        
        x = 0 
        if not np.array_equal(best, target): 
            x = 1
            sorted_pop = [x for x in sorted_pop if not np.array_equal(x[1][0], target)]
        indv = self._random_indv(len(sorted_pop) - x, 2, True)
        
        a = sorted_pop[indv[0]][1][0]
        b = sorted_pop[indv[1]][1][0]
            
        return target + self.f * (best - target) + self.f * (a - b)
```
